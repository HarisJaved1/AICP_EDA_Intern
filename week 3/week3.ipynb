```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Task 1


```python
# The data
x = [1.0, 2.0 , 3.0]
y = [2.0, 4.0 , 1.0]


# Making the plot
plt.plot(x, y)

# Naming labels and title
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Basic Graph 1")

# Applying the limits
plt.xlim(1, 3)
plt.ylim(1, 4)

#Showing the ticks on x axis
plt.xticks([1, 1.5, 2, 2.5, 3])

plt.show()
```

#### Task 2

```python
# The data
x= [10 , 20, 30]
ya = [20, 40, 10]
yb = [40, 10, 30]

#Creating the plot
L1 = plt.plot(x, ya, label= "Line 1", color="green", linewidth=2)
L2 = plt.plot(x, yb, label= "Line 2", color= "red", linewidth= 4)

# Displaying the legends 
plt.legend()

#Adding labels and title
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("Graph of two lines")

plt.show()
```

#### Task 3

```python
x= [1, 4, 5, 6, 7]
y= [2, 6, 3, 6, 3]
ya= [3, 7, 4, 7, 4]

plt.plot(x, y, linestyle= 'dotted', color='purple', linewidth= 3)
plt.plot(x, ya, linestyle= 'dashed', color = 'green', linewidth=3)

#Adding the x and y limits 
plt.xlim(1, 8)
plt.ylim(1, 8)


# Adding the markers
plt.scatter(x, y, marker="o", color='red')
plt.scatter(x, ya, marker='D')

#adding labels and title
plt.xlabel("x - axis")
plt.ylabel("y - label")
plt.title("Plot with two lines")

plt.show()

```

#### Task 6

```python
data = pd.read_csv('medals.csv')

top_5_countries = data.sort_values(by="goldmedals", ascending=False).head(5)

countries = top_5_countries["Country"].to_list()
gold_medals = top_5_countries["goldmedals"].to_list()
 
#adding the exploding effect 
explode = [0] * len(countries)  # All slices start at 0
explode[countries.index("United States")] = 0.1 

#Creating the pie chart
plt.pie(gold_medals, labels=countries, autopct="%1.1f%%", startangle=160, explode =explode, shadow=True)
plt.title("Top 5 Countries with Most Gold Medals in 2016 Olympics")
plt.axis('equal') 

plt.show()
```
